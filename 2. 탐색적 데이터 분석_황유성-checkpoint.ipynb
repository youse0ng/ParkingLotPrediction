```python
# 함수 만들기
import matplotlib.pyplot as plt
import seaborn as sns


def show_continuous_univariable(df:pd.DataFrame,category:str)->None:
    ''' DOCSTRING
    
    * Parameter:
        ** df: Pandas.DataFrame 분석을 원하는 데이터프레임 (데이터프레임 자료형)
        ** category: 분석을 원하는 칼럼명(String)
        Histogram과 Kdeplot, Boxplot을 그리고 저장합니다.
    '''
    plt.figure(figsize=(6.5,6.5))
    plt.subplot(2,1,1)
    sns.histplot(df,x=category,kde=True,palette=custom_palette)
    plt.title(f"{category}의 히스토그램과 Kdeplot")
    plt.xlabel(f'{category}')
    plt.ylabel('Count')
    
    plt.subplot(2,1,2)
    sns.boxplot(df,x=category,palette=custom_palette)
    plt.title(f"{category}의 BoxPlot")
    plt.xlabel(f'{category}')
    plt.ylabel('Count')
    
    plt.tight_layout()
    plt.title(f"{category}의 연속형 단변량 시각화")
    plt.savefig(f'{category}의 연속형 단변량 시각화.png')
    plt.show()
    
    
```

### (1) 숫자형 변수들 간의 상관관계

- 숫자형 변수 → 실차량수 상관관계를 분석합니다.    
- 숫자형 변수들 간의 상관관계를 df.corr() 메서드와 sns.heatmap() 함수 등을 사용해 시각화 합니다.

### (2) 숫자형 변수 → 실차량수

- 상관계수가 높은 상위 3~5개 숫자형 변수와 실차량수 관계를 산점도를 시각화해 분석합니다.
- 반복되는 코드는 함수를 만들어 사용하면 편리합니다.

```python
# 함수 만들기
def show_scatterplot(df,column:str,target:str='실차량수')->None:
    plt.figure(figsize=(7,7))
    df = df.loc[df[column]>0]
    sns.lmplot(df,x=column,y=target,markers='.',line_kws={'color':'red'},palette=custom_palette)
    plt.grid()
    r_estimate, pvalue = pearsonr(df[column],df[target])
    if pvalue < 0.05:
        print("귀무가설을 기각합니다. (즉, 대립가설을 채택 : 두 변수와의 상관관계가 존재합니다.)")
        print(f'P-value: {pvalue}입니다.')
    else:
        print("귀무가설을 채택합니다. (즉, 두 변수와의 상관관계가 존재하지 않습니다.)")
        print(f'P-value: {pvalue}입니다.')
    print(f'두 변수의 상관관계는 {r_estimate} 입니다.')
    plt.show()
```

### (3) 범주형 변수 → 실차량수

- 범줏값 별 실차량수 평균을 막대그래프로 시각화해 분석합니다.
- 반복되는 코드는 함수를 만들어 사용하면 편리합니다.

```python
# 함수 만들기
import scipy.stats as sm

def show_category(df:pd.DataFrame,column:str,hue:str=None,target='실차량수')->None:
    """ 범주형 변수와 실차량수와의 관계를 파악합니다.
        범주값별 실차량수의 평균을 막대그래프로 그립니다.
    Args:
        df (pd.DataFrame): 분석을 원하는 데이터프레임
        column (str): 분석을 원하는 칼럼
        target (_type_): 타겟 변수
        hue: string : 분석을 나누고 싶은것
    """
    plt.figure(figsize=(10,6))
    sns.barplot(df,x=column,y=target,hue=hue)
    plt.xticks(rotation=45)
    plt.show()
```

## 5. 추가 전처리

- '난방방식', '승강기설치여부' 두 변수에 대해서 적은 수의 범주로 묶는 작업을 진행합니다.
- '단지코드'는 개별적인 값이므로 예측과 관련이 없으니 제거합니다.
- '지역' 변수는 지역별 특성을 고려하기에는 너무 세분화 되어 있고, 특징으로 통합하기 어려우므로 제거합니다.

### (1) 난방방식

- '난방방식' 변수의 값을 다음과 같이 변경하여 세 개의 범줏값만 갖게 합니다.
    - 개별가스난방 → 개별
    - 개별유류난방 → 개별
    - 지역난방 → 지역
    - 지역가스난방 → 지역
    - 지역유류난방 → 지역
    - 중앙가스난방 → 중앙
    - 중앙난방 → 중앙
    - 중앙유류난방 → 중앙

```python
apart['난방방식'] = apart['난방방식'].replace({'개별가스난방':'개별','개별유류난방':'개별','지역난방':'지역','지역가스난방':'지역','지역유류난방':'지역','중앙가스난방':'중앙','중앙난방':'중앙','중앙유류난방':'중앙'})
apart
```

### (2) 승강기설치여부

- '승강기설치여부' 변수의 값을 다음과 같이 변경하여 0, 1 값을 갖게 합니다.
    - 전체동 설치 → 1
    - 일부동 설치 → 0
    - 미설치 → 0

```python
apart['승강기설치여부'] = apart['승강기설치여부'].replace({'전체동 설치':1,'일부동 설치':0,'미설치':0})
apart
```

### (3) 단지코드, 지역 제거

- '단지코드', '지역' 변수를 제거합니다.

```python
drop_cols = ['단지코드','지역']

apart.drop(columns=drop_cols,inplace=True)
```
