```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns

import joblib
import warnings

warnings.filterwarnings(action='ignore')
%config InlineBackend.figure_format='retina'
```

# 난방 방식에 대한 결측치를 KNN으로 예측을 해야함

```python
apart = pd.read_excel(''+'train.xlsx')
apart
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
drop_cols = ['단지명','단지코드','지역','건물형태','승강기설치여부'] # 범주형 칼럼은 가변수화에 있어 문제가 생겨 아예 삭제하는 방향으로 잡음
apart = apart.drop(columns=drop_cols)
apart
```

# y_test는 결측치가 존재하지 않는 난방방식에서 y_test를 뽑아냄

```python
y_test = apart.loc[apart['난방방식'].isnull()]
y_test = y_test['난방방식']
```

# 준공년도 전처리

```python
apart['준공년도'] = apart['준공일자']//10000
apart.drop(columns='준공일자',inplace=True)
apart
```

# x_test 추출

```python
x_test = apart.loc[apart['난방방식'].isnull()] # 건물형태가 어떨 것이다 라고 예측해야할 데이터
x_test = x_test.drop(columns='난방방식')
x_cols = list(x_test)
x_test
```

# 학습할 x_train과 y_train을 추출하기 (결측치는 존재하지 않아야함) | x_train,x_test 정규화 작업

```python
scaler = MinMaxScaler()

apart=apart.dropna(subset=['난방방식'])

target = '난방방식'
x_train = apart.drop(columns=target)
y_train = apart['난방방식']

x_train = scaler.fit_transform(x_train)
x_train = pd.DataFrame(x_train,columns=x_cols)

x_test = scaler.transform(x_test)
x_test = pd.DataFrame(x_test,columns=x_cols)
x_test
```

```python
display(x_train.head())
display(x_test.head())
display(y_train.isnull().sum())
print(x_train.shape,y_train.shape)
print(y_train)
```

```python
model = KNeighborsClassifier(3)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

# 결측치에 대한 예측값 추출

```python
y_pred = model.predict(x_test)
```

```python
y_test = pd.DataFrame(y_test)
y_test['난방방식'] = y_pred
```
